```python
import torch 
from torch_geometric.utils import degree 
import pickle 
from collections import Counter 
```

```python
with open('./processed_data/graph.pkl', 'rb') as r:
    graph = pickle.load(r) 

graph 
```

```python
Counter(graph.y.tolist())
```

```python
degree_1d = degree(
    index = graph.edge_index[0],
    num_nodes = graph.num_nodes,
    dtype = torch.int64,
)

(degree_1d == 0).sum(), degree_1d.float().mean(), degree_1d.min(), degree_1d.max() 
```

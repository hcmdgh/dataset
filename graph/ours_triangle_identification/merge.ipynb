```python
import torch 
from torch_geometric.data import Batch 
import pickle 
```

```python
with open('./processed_data/train_val_test_graph.pkl', 'rb') as r:
    train_graph, val_graph, test_graph = pickle.load(r)

train_graph, val_graph, test_graph
```

```python
num_nodes = int(train_graph.num_nodes)
ones = torch.ones(num_nodes, dtype=torch.bool)
zeros = torch.zeros(num_nodes, dtype=torch.bool)

train_graph.train_mask = ones 
train_graph.val_mask = zeros
train_graph.test_mask = zeros
val_graph.train_mask = zeros  
val_graph.val_mask = ones 
val_graph.test_mask = zeros
test_graph.train_mask = zeros  
test_graph.val_mask = zeros 
test_graph.test_mask = ones

graph = Batch.from_data_list([train_graph, val_graph, test_graph])

graph 
```

```python
with open('./processed_data/graph.pkl', 'wb') as w:
    pickle.dump(graph, w)
```
